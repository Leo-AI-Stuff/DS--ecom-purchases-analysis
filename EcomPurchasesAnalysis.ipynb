```python
#print data where language is french
dataset[dataset['language'] == 'fr']
```

```python
#show jobs containing engineer keyword 
dataset[dataset['job'].str.contains('engineer', case=False)]['job']
```

```python
#find email of the guy with given ip address
dataset[dataset['ip address'] == '132.207.160.22']['email']
```

```python
#check how many people made purchase over 50 using master card
mastercardUsersWithAbove50Purchase = dataset[(dataset['cc provider'].str.contains('mastercard', case=False)) & (dataset['purchase price'] > 50)]
len(mastercardUsersWithAbove50Purchase)
```
